```python
from nltk.tokenize import word_tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
x = dataset.drop(['full_text'], axis=1)
y = dataset['full_text']
y
```

```python
# Using Tokenization

from nltk import tokenize

scores = []
sia = SentimentIntensityAnalyzer()

for text in y:
#     if len(text) > 100: # If tweet length > 100
    # Need to tokenize
        text_sublist = tokenize.sent_tokenize(text) # Breakdown into lists
        text_sentiment = 0.0

        for subtext in text_sublist: # Traverse through each list item
            score = sia.polarity_scores(subtext)
            text_sentiment += score['compound'] # Add compound score of each list item
            text_sentiment = round(text_sentiment / len(text_sublist), 2) # Round Average upto 2 DP. 

        scores.append(text_sentiment) # Append average score to sentiment_vader

#     else:
#         scores.append(sia.polarity_scores(text))
scores
    # print(str(round(text_sentiment / len(text_sublist), 2)))
```

```python
df = pd.DataFrame({'full_text': y, 'sentiment_vader': scores})
df
```

```python
# Normalisation If Tokenised

sentiment_vader = []

for score in scores:
#     if(type(score)==dict):
#         if score['compound'] >= 0.5:
#             sentiment_vader.append(1)
#         elif score['compound'] <= -0.5:
#             sentiment_vader.append(-1)
#         else:
#             sentiment_vader.append(0)

#         if score['pos'] > score['neg'] and score['pos'] > score['neu']:
#             sentiment_vader.append(1)
#         elif score['neg'] > score['pos'] and score['neg'] > score['neu']:
#             sentiment_vader.append(-1)
#         else:
#             sentiment_vader.append(0)
#     else:
        if score >= 0.2:
            sentiment_vader.append(1)
        elif score <= -0.3:
            sentiment_vader.append(-1)
        else:
            sentiment_vader.append(0)
        
sentiment_vader
```

```python
df['sentiments_vader'] = sentiment_vader
```

```python
df[(df['sentiments_vader']>-0.05) & (df['sentiments_vader']<0.05)]
```

```python
dataset = dataset.assign(sentiments_vader = df['sentiments_vader'].values)
dataset
```
